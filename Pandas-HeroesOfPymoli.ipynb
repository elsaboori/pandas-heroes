Age Demographics


```python

bins = [0,10,14,19,24,29,34,39,100]
group_name = ["<10", "10-14","15-19", "20-24", "25-29","30-34", "35-39", "40+" ]
heroes_df["Age Range"] = pd.cut(heroes_df["Age"], bins, labels=group_name)
age_range_grouped = heroes_df[["Age Range","Gender"]].groupby(["Age Range"]).count()
Total_count = age_range_grouped["Gender"].sum()
age_range_grouped["Total Count"] = age_range_grouped["Gender"]*100/Total_count
age_range_grouped = age_range_grouped.rename(columns={"Gender":"Total Count", "Total Count":"Percentage of Players"})
age_range_grouped
```

Average Purchase Price


```python

heroes_grouped =heroes_df[["Age Range", "SN", "Price"]].groupby(["Age Range"]).agg({"Price" : ["count", "mean", "sum"]})
heroes_grouped.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
heroes_grouped["Normalized Totals"] = heroes_grouped["Total Purchase Value"]/heroes_grouped["Total Purchase Value"].sum()
heroes_grouped
```

Top Spenders


```python

Top_Spenders = heroes_df[["SN", "Price"]].groupby(["SN"]).agg({"Price":["count","mean","sum"]})
Top_Spenders.columns=["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
Top_Spenders.sort_values("Total Purchase Value" , ascending=False).head(5)
```

Most Popular Items


```python

Most_Popilar_Item = heroes_df[["Item ID","Item Name","Price"]].groupby(["Item ID", "Item Name", "Price"]).agg({"Price":["count", "sum"]})
Most_Popilar_Item.columns=["Purchase Count", "Total Purchase Value"]
Most_Popilar_Item.sort_values("Purchase Count", ascending=False).head(5)
```
